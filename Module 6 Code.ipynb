```python
import pandas as pd
import numpy as np

from collections import Counter
```

```python
diabetes_data = pd.read_csv("C:/Users/bahla/Desktop/INST414/diabetes_data.csv")
```

```python
print(diabetes_data.head(60))
```

```python
patient_feature_map = {}
patient_risk_map = {}

for idx, row in diabetes_data.iterrows():
    patient_id = row["user_id"]
    patient_features = row.drop(["user_id", "date", "risk_score"])
    patient_feature_map[patient_id] = patient_features
    patient_risk_map[patient_id] = row["risk_score"]
```

```python
similarity_scores = []
target_patient_id = list(patient_feature_map.keys())[0]
```

```python
for patient_id, features in patient_feature_map.items():
    if patient_id == target_patient_id:
        continue
```

```python
target_features = patient_feature_map[target_patient_id]
intersection = np.sum((target_features == features) & target_features.isnull())
union = np.sum((target_features != 0) | (features != 0))
jaccard_similarity = intersection / union if union != 0 else 0

similarity_scores.append({"Patient ID": patient_id, "Jaccard Similarity": jaccard_similarity})
```

```python
similarity_scores_df = pd.DataFrame(similarity_scores)
```

```python
k_nn = 5
similarity_scores_df_sorted = similarity_scores_df.sort_values(by = "Patient ID", ascending = False).head(k_nn)
top_k_ids = similarity_scores_df_sorted["Patient ID"].tolist()
```

```python
risk_categories = [patient_risk_map[pid] for pid in top_k_ids]
most_common_risk = Counter(risk_categories).most_common(1)[0][0]
```

```python
print(f"Most common risk categories among top {k_nn} similar patients: {most_common_risk}")
```

```python
average_risk_score = np.mean([patient_risk_map[pid] for pid in top_k_ids])
print(f"Average risk score of top {k_nn} similar patients: {average_risk_score}")
```
